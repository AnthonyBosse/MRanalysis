# Mohn Ridge Frontal Current glider transport analysis (Bosse and Fer, GRL 2019)

This package presents the transport analysis and results of glider data collected across the Mohn Ridge (Nordic Seas).  
It intends to provide a way to repdroduce the results published in [*Bosse, A., & Fer, I. (2019). Mean structure and seasonality of the Norwegian Atlantic Front Current along the Mohn Ridge from repeated glider transects. Geophysical Research Letters, 46(22), 13170-13179.*](https://doi.org/10.1029/2019GL084723), but also enable users to apply the method to other glider data sets aiming at describing geostrophic currents and transport.  
This notebook was written in Matlab and can be run with Octave, an open-source alternative to Matlab.

## Computing geostrophic velocities

This is a now crucial step where geostrophic velocities are computed. It follows several steps :  
- Calculation of frontal direction given by DAC and angle between the glider track and average current.
- Gridding and smootinh of the T,S sections prior to gradient calculations
- Geostrophic shear is obtained by differentiating the dynamic height and referenced by DAC. 
- The angle between DAC and glider track is used to correct the geostrophic velocities, vorticity and distance.

```matlab
% compute Vg and other PV diagnosticis (this step might take a bit of time)
[glid_pv1] = comp_runningPV(sg560,5);[glid_pv2] = comp_runningPV(sg562,5);
```

```matlab
% Merge glid_pv1 and glid_pv2
varr = fieldnames(glid_pv1);
for ll=3:length(varr)
    eval(['glid_pv.' varr{ll} ' = [glid_pv1.' varr{ll} ' glid_pv2.' varr{ll} '];'])
end
glid_pv.d_along = [glid_pv1.d_along  glid_pv1.d_along(end)+glid_pv2.d_along];
glid_pv.d_cor = nancumsum([0 diff(nanmean(glid_pv.d_along)).*red_x(nanmean(glid_pv.sint))]);

glid_pv
```

The resulting structure contains the following fields for each glider profile:  
- **dir_flow** : the direction of the flow, ie given by the DAC smoothed
- **pp** and **d_along** : pressure and cross-track distance 
- **CTs**, **SAs**, **SIGs** : low-pass filtered temperature, salinity and density sections gridded on (pp,d_along)
- **sint** : sinus of angle between DAC and glider track used to get cross-tream velocities (see Appendix of Bosse and Fer (2019) for a schematics)
- **v_tot**, **ff**, **dBdx**, **vort**, **pv_A**, **pv_B**, **pv** (*all fields with "_cor" are geometricaly corrected  transformation to . *)
- **d_cor** : the cumulated cross-stream distance along the glider track

Now the idea is to compute the same anaylsis for individual sections. It follows different steps :  
- optimal interpolation of T,S in along-stream distance coordinate system (correlation length of $Lsmo$ and $Zsmo$ chosen at 20km and 15m as in Bosse and Fer (2019), with y=0 defined at the velocity maximum
- vertical shear is obtained from the resulting T,S cross-front distribution and referenced using DAC intensity
- Vg is divided into a geostrophic part (shear integrated above 1000m) and a deep component part (mean absolute geostrophic current between 950-1000m)
- compute different properties such as total transport, AW transport etc  
(This step can take a bit of time, in particular with the Optimal interpolation step)

```matlab
% get along-track distance and along-stream distance (perpendicular to DAC)
d_along = nanmean(glid_pv.d_along); dc_along = glid_pv.d_cor;
reload_io = 0; % set to 1 to force reloading optimal interpolation
zz = dir('data');
clear glid_MR TR
if ~isempty(find(strcmp({zz.name},'glid_MR.mat')))
    load('data/glid_MR.mat')
else
    reload_io = 1; % if no matfile then force to reload optimal interpolation
end

for ll= 1:length(sec_beg)]
    opt_vis = 0;
    glid_MR.Lsmo = 20;glid_MR.Zsmo = 15;glid_MR.err = 0.05; % correlation length scale for OI
    dr = 2.5;dz = 5; % grid resolution for optimal ineterpolation 
    comp_glid_section_MR % do the MR front analysis for an individual section
    plot_glid_section_MR % plot some figures % skip to process faster
end
% save results to avoid recomputing everytime
save('data/glid_MR.mat','glid_MR','TR');
```
